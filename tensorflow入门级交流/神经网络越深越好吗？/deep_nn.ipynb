# 用神经网络的方法来拟合三角函数的数据

## 一、制造数据
加入噪声

```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt

x_data = np.linspace(-10,10,2000,dtype=np.float32)[:, np.newaxis]
y_noise = np.random.normal(0,0.1,x_data.shape)
y_data = np.sin(x_data) + y_noise
print(np.shape(x_data))
plt.plot(x_data,y_data)
plt.show()
```

## 二、构建BP神经网络

```python
def weight_variable(shape):
    initial = tf.truncated_normal(shape, stddev=0.1)
    return tf.Variable(initial)

def bias_variable(shape):
    initial = tf.constant(0.1, shape=shape)
    return tf.Variable(initial)

#占位
x = tf.placeholder(tf.float32, [None, 1],name='x')
y = tf.placeholder(tf.float32, [None, 1],name='y')
keep_prob = tf.placeholder(tf.float32,name='keep_prob')
#use `rate` instead of `keep_prob`. Rate should be set to `rate = 1 - keep_prob`.
#rate = tf.placeholder(tf.float32,name='rate')

#first fully layer
w_fc1 = weight_variable([1,128])
b_fc1 = bias_variable([128])
h_fc1 = tf.nn.relu(tf.nn.bias_add(tf.matmul(x, w_fc1), b_fc1))
h_fc1_drop = tf.nn.dropout(h_fc1, keep_prob)

#2nd layer
w_fc2 = weight_variable([128,128])
b_fc2 = bias_variable([128])
h_fc2 = tf.nn.relu(tf.nn.bias_add(tf.matmul(h_fc1_drop, w_fc2), b_fc2))
h_fc2_drop = tf.nn.dropout(h_fc2, keep_prob)

#3nd layer
w_fc3 = weight_variable([128,128])
b_fc3 = bias_variable([128])
h_fc3 = tf.nn.relu(tf.nn.bias_add(tf.matmul(h_fc2_drop, w_fc3), b_fc3))
h_fc3_drop = tf.nn.dropout(h_fc3, keep_prob)

#4nd layer
w_fc4 = weight_variable([128,128])
b_fc4 = bias_variable([128])
h_fc4 = tf.nn.relu(tf.nn.bias_add(tf.matmul(h_fc3_drop, w_fc4), b_fc4))
h_fc4_drop = tf.nn.dropout(h_fc4, keep_prob)

#5nd layer
w_fc5 = weight_variable([128,128])
b_fc5 = bias_variable([128])
h_fc5 = tf.nn.relu(tf.nn.bias_add(tf.matmul(h_fc4_drop, w_fc5), b_fc5))
h_fc5_drop = tf.nn.dropout(h_fc5, keep_prob)

#6nd layer
w_fc6 = weight_variable([128,128])
b_fc6 = bias_variable([128])
h_fc6 = tf.nn.relu(tf.nn.bias_add(tf.matmul(h_fc5_drop, w_fc6), b_fc6))
h_fc6_drop = tf.nn.dropout(h_fc6, keep_prob)

#7nd layer
w_fc7 = weight_variable([128,128])
b_fc7 = bias_variable([128])
h_fc7 = tf.nn.relu(tf.nn.bias_add(tf.matmul(h_fc6_drop, w_fc7), b_fc7))
h_fc7_drop = tf.nn.dropout(h_fc7, keep_prob)

#8nd layer
w_fc8 = weight_variable([128,128])
b_fc8 = bias_variable([128])
h_fc8 = tf.nn.relu(tf.nn.bias_add(tf.matmul(h_fc7_drop, w_fc8), b_fc8))
h_fc8_drop = tf.nn.dropout(h_fc8, keep_prob)

#9nd layer
w_fc9 = weight_variable([128,128])
b_fc9 = bias_variable([128])
h_fc9 = tf.nn.relu(tf.nn.bias_add(tf.matmul(h_fc8_drop, w_fc9), b_fc9))
h_fc9_drop = tf.nn.dropout(h_fc9, keep_prob)

#10nd layer
w_fc10 = weight_variable([128,128])
b_fc10 = bias_variable([128])
h_fc10 = tf.nn.relu(tf.nn.bias_add(tf.matmul(h_fc9_drop, w_fc10), b_fc10))
h_fc10_drop = tf.nn.dropout(h_fc10, keep_prob)

#11 layer
w_fc11 = weight_variable([128,128])
b_fc11 = bias_variable([128])
h_fc11 = tf.nn.relu(tf.nn.bias_add(tf.matmul(h_fc10_drop, w_fc11), b_fc11))
h_fc11_drop = tf.nn.dropout(h_fc11, keep_prob)

#12nd layer
w_fc12 = weight_variable([128,128])
b_fc12 = bias_variable([128])
h_fc12 = tf.nn.relu(tf.nn.bias_add(tf.matmul(h_fc11_drop, w_fc12), b_fc12))
h_fc12_drop = tf.nn.dropout(h_fc12, keep_prob)

#13nd layer
w_fc13 = weight_variable([128,128])
b_fc13 = bias_variable([128])
h_fc13 = tf.nn.relu(tf.nn.bias_add(tf.matmul(h_fc12_drop, w_fc13), b_fc13))
h_fc13_drop = tf.nn.dropout(h_fc13, keep_prob)

#14nd layer
w_fc14 = weight_variable([128,128])
b_fc14 = bias_variable([128])
h_fc14 = tf.nn.relu(tf.nn.bias_add(tf.matmul(h_fc13_drop, w_fc14), b_fc14))
h_fc14_drop = tf.nn.dropout(h_fc14, keep_prob)

#15nd layer
w_fc15 = weight_variable([128,128])
b_fc15 = bias_variable([128])
h_fc15 = tf.nn.relu(tf.nn.bias_add(tf.matmul(h_fc14_drop, w_fc15), b_fc15))
h_fc15_drop = tf.nn.dropout(h_fc15, keep_prob)

#16nd layer
w_fc16 = weight_variable([128,128])
b_fc16 = bias_variable([128])
h_fc16 = tf.nn.relu(tf.nn.bias_add(tf.matmul(h_fc15_drop, w_fc16), b_fc16))
h_fc16_drop = tf.nn.dropout(h_fc16, keep_prob)

#17nd layer
w_fc17 = weight_variable([128,128])
b_fc17 = bias_variable([128])
h_fc17 = tf.nn.relu(tf.nn.bias_add(tf.matmul(h_fc16_drop, w_fc17), b_fc17))
h_fc17_drop = tf.nn.dropout(h_fc17, keep_prob)

#18nd layer
w_fc18 = weight_variable([128,128])
b_fc18 = bias_variable([128])
h_fc18 = tf.nn.relu(tf.nn.bias_add(tf.matmul(h_fc17_drop, w_fc18), b_fc18))
h_fc18_drop = tf.nn.dropout(h_fc18, keep_prob)

#19nd layer
w_fc19 = weight_variable([128,128])
b_fc19 = bias_variable([128])
h_fc19 = tf.nn.relu(tf.nn.bias_add(tf.matmul(h_fc18_drop, w_fc19), b_fc19))
h_fc19_drop = tf.nn.dropout(h_fc19, keep_prob)

#20 layer
w_fc20 = weight_variable([128,1])
b_fc20 = bias_variable([1])
y_pred = tf.nn.bias_add(tf.matmul(h_fc19_drop, w_fc20), b_fc20,name='y_pred')
```

## 三、定义损失函数和优化器

```python
loss = tf.reduce_mean(tf.square(y - y_pred))
train_step = tf.compat.v1.train.AdamOptimizer(1e-4).minimize(loss)
```

## 四、开始运算

```python
with tf.Session() as sess:
    sess.run(tf.compat.v1.global_variables_initializer())
    step = 1
    train_loss=1
    while step<=10001:
        sess.run(train_step,feed_dict={x: x_data, y: y_data, keep_prob: 0.75})
        if step%100 == 0:
            pred,train_loss = sess.run([y_pred,loss],feed_dict={x: x_data, y: y_data, keep_prob: 1.00})
            print ('step:%d,train loss:%f' % (step,train_loss))
            plt.plot(x_data,y_data,'b',x_data,pred,'r')
            plt.show()
        step += 1
```
